```python
import cv2 # Baris ini mengimpor modul cv2 dari library OpenCV yang digunakan untuk pemrosesan gambar dan komputer visi.
import numpy as np # Baris ini mengimpor modul numpy yang merupakan modul yang digunakan untuk melakukan operasi numerik dan manipulasi array dalam Python.
from skimage import io # Baris ini mengimpor modul io dari library skimage yang digunakan untuk membaca dan menulis gambar dalam berbagai format.
import matplotlib.pyplot as plt # Baris ini mengimpor modul pyplot dari library matplotlib yang digunakan untuk membuat plot dan visualisasi data.

img = io.imread('yelena.jpg') # Baris ini membaca citra dengan nama file 'yelena.jpg' menggunakan fungsi io.imread() dari modul skimage. Citra tersebut disimpan dalam variabel img.

row, column, _ = img.shape # aris ini mendapatkan dimensi baris (row) dan kolom (column) dari citra img menggunakan atribut .shape. Tanda garis bawah terakhir _ digunakan untuk mengabaikan dimensi ketiga (channel) dari citra, karena dalam contoh ini citra adalah citra berwarna.

img1 = np.zeros((row, column), dtype='uint8') # Baris ini membuat array numpy berukuran (row, column) yang diinisialisasi dengan nol, dengan tipe data uint8 (unsigned integer 8-bit). Array ini akan digunakan untuk menyimpan citra hasil pemrosesan.

min_range = 10 # Baris ini menentukan rentang nilai intensitas piksel yang akan digunakan dalam proses pemrosesan.
max_range = 60 # Nilai piksel di luar rentang ini akan diubah menjadi nol (hitam), sedangkan piksel di dalam rentang akan diubah menjadi 255 (putih).

for i in range(row): # Baris ini melakukan iterasi pada setiap piksel citra img. Jika nilai intensitas piksel pada saluran biru (img[i, j, 0]) berada di antara min_range dan max_range, 
    for j in range(column): # maka piksel pada citra hasil (img1) diubah menjadi 255 (putih), jika tidak, diubah menjadi 0 (hitam).
        if img[i, j, 0] > min_range and img[i, j, 0] < max_range:
            img1[i, j] = 255
        else:
            img1[i, j] = 0

fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # Baris ini membuat objek gambar (Figure) dan array objek sumbu (AxesSubplot) dengan ukuran 2x2, yang akan menampilkan 4 subplot dalam grid 2x2. Ukuran keseluruhan gambar ditentukan oleh figsize=(12, 12) dengan lebar dan tinggi masing-masing 12 inci.
ax = axes.ravel() # Baris ini meratakan array axes menjadi array satu dimensi menggunakan metode ravel(). Variabel ax akan menyimpan array satu dimensi yang berisi objek sumbu untuk setiap subplot.

ax[0].imshow(img, cmap=plt.cm.gray) # Baris ini menampilkan citra input (img) pada subplot pertama (ax[0]) dengan menggunakan imshow() dari objek sumbu. Parameter cmap=plt.cm.gray digunakan untuk menampilkan citra dalam skala abu-abu (grayscale).
ax[0].set_title("Citra Input") # set_title() digunakan untuk mengatur judul subplot menjadi "Citra Input".
ax[1].hist(img[:, :, 0].ravel(), bins=256) # Baris ini menampilkan histogram citra input pada subplot kedua (ax[1]) menggunakan hist() dari objek sumbu. img[:, :, 0] digunakan untuk mengakses saluran biru dari citra input. ravel() digunakan untuk meratakan array histogram menjadi satu dimensi. Parameter bins=256 digunakan untuk menentukan jumlah bin dalam histogram.
ax[1].set_title('Histogram Input') # set_title() digunakan untuk mengatur judul subplot menjadi "Histogram Input".

ax[2].imshow(img1, cmap=plt.cm.gray) # Baris ini menampilkan citra output (img1) pada subplot ketiga (ax[2]) dengan menggunakan imshow(). Parameter cmap=plt.cm.gray digunakan untuk menampilkan citra dalam skala abu-abu.
ax[2].set_title("Citra Output") # set_title() digunakan untuk mengatur judul subplot menjadi "Citra Output".
ax[3].hist(img1.ravel(), bins=256) # Baris ini menampilkan histogram citra output pada subplot keempat (ax[3]) menggunakan hist(). img1.ravel() digunakan untuk meratakan array citra output menjadi satu dimensi. Parameter bins=256 digunakan untuk menentukan jumlah bin dalam histogram.
ax[3].set_title('Histogram Output') # set_title() digunakan untuk mengatur judul subplot menjadi "Histogram Output".

plt.tight_layout() # Baris ini mengatur tata letak subplot agar lebih rapi dengan tight_layout().
plt.show() # show() digunakan untuk menampilkan gambar dan histogram dalam bentuk subplot yang telah dibuat.
```
