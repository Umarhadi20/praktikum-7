```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Membaca citra
img = cv2.imread('16401178.jpg')

# Konversi warna citra
cat = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Menampilkan citra asli
plt.subplot(1, 3, 1)
plt.imshow(cat)
plt.title('Original Image')
plt.axis('off')

# Filter dengan kernel rata-rata
kernel = np.ones((5, 5), np.float32) / 25
print(kernel)
kucing_filter = cv2.filter2D(img, -1, kernel)

# Menampilkan citra setelah filtering
plt.subplot(1, 3, 2)
plt.imshow(kucing_filter)
plt.title('Filtered Image')
plt.axis('off')

# Menampilkan histogram citra asli
plt.subplot(1, 3, 3)
plt.hist(cat.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Mengatur ukuran tampilan plot
plt.rcParams["figure.figsize"] = (12, 4)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15,15) # Baris ini mengatur ukuran gambar yang akan ditampilkan menggunakan plt.rcParams. Di sini, ukuran gambar diatur menjadi 15 inci x 15 inci.

plt.subplot(2, 2, 1)
plt.imshow(cat)
plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2)
plt.hist(cat.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Original Histogram')

# Menampilkan citra setelah filtering beserta histogram
plt.subplot(2, 2, 3)
plt.imshow(kucing_filter)
plt.title('Averaging')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 4)
plt.hist(kucing_filter.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Filtered Histogram')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
import cv2 # memanggil library opencv
import numpy as np # memanggil library numpy
from matplotlib import pyplot as plt # memanggil library matplotlib


img = cv2.imread('number.jpg',0) # membaca file

laplacian = cv2.Laplacian(img,cv2.CV_64F) # Baris ini menggunakan fungsi Laplacian dari modul OpenCV (cv2) untuk menerapkan operator Laplacian pada citra img. Operator Laplacian digunakan untuk mendeteksi perubahan intensitas citra di sekitar setiap piksel. Parameter kedua cv2.CV_64F menunjukkan tipe data output yang diinginkan.

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) # Baris ini menggunakan fungsi Sobel dari modul OpenCV (cv2) untuk menerapkan operator Sobel pada citra img dalam arah sumbu x (horizontal). Operator Sobel digunakan untuk mendeteksi tepi horisontal pada citra. Parameter ketiga 1 menunjukkan bahwa operator Sobel diterapkan pada sumbu x, sedangkan parameter keempat 0 menunjukkan bahwa operator Sobel tidak diterapkan pada sumbu y. Parameter terakhir ksize=5 menunjukkan ukuran kernel Sobel yang digunakan.
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5) # Baris ini mirip dengan baris ke-2, tetapi kali ini operator Sobel diterapkan pada sumbu y (vertikal) untuk mendeteksi tepi vertikal pada citra.

plt.rcParams["figure.figsize"] = (20,20) # Baris ini mengatur ukuran gambar yang akan ditampilkan menggunakan plt.rcParams. Di sini, ukuran gambar diatur menjadi 20 inci x 20 inci.

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 3)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 4)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y')
plt.xticks([]), plt.yticks([])

# Menampilkan histogram untuk setiap citra
plt.figure(figsize=(20, 10))

plt.subplot(2, 2, 1)
plt.hist(img.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Original Histogram')

plt.subplot(2, 2, 2)
plt.hist(laplacian.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Laplacian Histogram')

plt.subplot(2, 2, 3)
plt.hist(sobelx.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Sobel X Histogram')

plt.subplot(2, 2, 4)
plt.hist(sobely.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Sobel Y Histogram')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Membaca citra dalam skala keabuan (grayscale)
img = cv2.imread('hiro hamada.jpg', 0)

# Deteksi tepi menggunakan metode Canny
edges = cv2.Canny(img, 100, 200)

# Menampilkan plot
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.xticks([]), plt.yticks([])

# Menampilkan histogram untuk citra asli
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(img.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Original Image Histogram')

# Menampilkan histogram untuk citra tepi (edges)
plt.subplot(1, 2, 2)
plt.hist(edges.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Edge Image Histogram')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Membaca gambar baymax dalam skala keabuan (grayscale)
img = cv2.imread('hiro hamada.jpg', 0)

# Hitungan threshold
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

# Menampilkan hasil
titles = ['Gambar asli', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

# Menampilkan beberapa gambar sekaligus
plt.figure(figsize=(10, 12))
for i in range(6):
    plt.subplot(3, 4, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

    # Menampilkan histogram
    plt.subplot(3, 4, i+7)
    plt.hist(images[i].ravel(), bins=256, color='blue', alpha=0.7)
    plt.title('Histogram')
    plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Membaca gambar baymax dalam skala keabuan (grayscale)
img = cv2.imread('hiro hamada.jpg', 0)

# Menggunakan median blur untuk menghaluskan tepi objek pada citra
# Ini diperlukan agar thresholding memberikan hasil yang lebih baik
img = cv2.medianBlur(img, 5)

# Lakukan Thresholding
# Binary Threshold
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Adaptive Threshold dengan Mean
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                            cv2.THRESH_BINARY, 11, 2)

# Adaptive Threshold dengan Gaussian
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv2.THRESH_BINARY, 11, 2)

# Plotting
titles = ['Original Image', 'Global Thresholding (v = 127)',
          'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

# Menampilkan hasil
plt.figure(figsize=(10, 8))
for i in range(4):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

    # Menampilkan histogram
    plt.subplot(2, 4, i+5)
    plt.hist(images[i].ravel(), bins=256, color='blue', alpha=0.7)
    plt.title('Histogram')
    plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
