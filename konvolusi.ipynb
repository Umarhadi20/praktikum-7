```python
import matplotlib.pyplot as plt # memanggil library matplotlib.pyplot
%matplotlib inline

from skimage import data # memanggil library skimage
from skimage.io import imread # memanggil library skimage
from skimage.color import rgb2gray # memanggil library skimage
import numpy as np  # memanggil library numpy
import cv2 # memanggil library opencv

citra1 = imread(fname="13074004.jpg") # membaca file
print(citra1.shape) # mencetak bentuk atau dimensi citra 1

plt.imshow(citra1, cmap='gray') # menampilkan gambar citra 1
```

```python
kernel = np.array([[-1, 0, -1], # Baris ini membuat sebuah kernel atau filter dengan menggunakan np.array dari NumPy.
                   [0, 4, 0],  # merupakan sebuah matriks 3x3
                   [-1, 0, -1]])

citraOutput = cv2.filter2D(citra1, -1, kernel) # Baris ini menerapkan operasi konvolusi pada citra1 menggunakan kernel yang telah dibuat sebelumnya. 

fig, axes = plt.subplots(1, 2, figsize=(12, 12)) # membuat dua plot satu baris dengan ukuran gambar 12x12
ax = axes.ravel() # mengubah array dua dimensi menjadi array satu dimensi

ax[0].imshow(citra1, cmap = 'gray') # menampilkan citra1
ax[0].set_title("Citra Input") # judul citra input
ax[1].imshow(citraOutput, cmap = 'gray') # menampilkan output citra1
ax[1].set_title("Citra Output") # judul citra output
```
