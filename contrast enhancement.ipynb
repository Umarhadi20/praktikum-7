```python
import numpy as np # Baris ini mengimpor modul numpy yang merupakan modul yang digunakan untuk melakukan operasi numerik dan manipulasi array dalam Python.
import matplotlib.pyplot as plt # Baris ini mengimpor modul pyplot dari library matplotlib yang digunakan untuk membuat plot dan visualisasi data.
import cv2 # Baris ini mengimpor modul cv2 dari library OpenCV yang digunakan untuk pemrosesan gambar dan komputer visi.
import matplotlib.image as mpimg # Baris ini mengimpor modul image dari library matplotlib yang digunakan untuk membaca dan memanipulasi gambar.
from skimage import data # Baris ini mengimpor modul data dari library skimage yang berisi beberapa contoh gambar yang bisa digunakan untuk keperluan pengujian dan pengembangan algoritma pemrosesan gambar.
from skimage import io # Baris ini mengimpor modul io dari library skimage yang digunakan untuk membaca dan menulis gambar dalam berbagai format.

image = io.imread('hatice.jpg') # Baris ini membaca gambar dengan nama file 'hatice.jpg' menggunakan fungsi io.imread() dari modul skimage. Gambar tersebut akan disimpan dalam variabel image.

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Baris ini mengonversi gambar ke citra skala abu-abu menggunakan fungsi cvtColor() dari modul cv2 (OpenCV). Parameter cv2.COLOR_BGR2GRAY mengindikasikan bahwa gambar akan diubah ke citra skala abu-abu dari format BGR.

gray_image_equalized = cv2.equalizeHist(gray_image) # Baris ini melakukan histogram equalization pada citra skala abu-abu menggunakan fungsi equalizeHist() dari modul cv2 (OpenCV). Hasilnya akan disimpan dalam variabel gray_image_equalized.

clahe = cv2.createCLAHE(clipLimit=2, tileGridSize=(8, 8)) # aris ini membuat objek CLAHE (Contrast Limited Adaptive Histogram Equalization) dengan batasan klip sebesar 2 dan ukuran grid 8x8 menggunakan fungsi createCLAHE() dari modul cv2 (OpenCV).
clahe_image = clahe.apply(gray_image) # metode apply() digunakan untuk menerapkan CLAHE pada citra skala abu-abu gray_image. Hasilnya akan disimpan dalam variabel clahe_image.

image_cs = np.zeros(gray_image.shape, dtype='uint8') # Baris ini membuat array nol dengan ukuran yang sama dengan citra skala abu-abu gray_image menggunakan fungsi np.zeros(). Tipe data array ditentukan sebagai uint8.

min_val = np.min(gray_image) # Baris ini menghitung nilai minimum dari citra skala abu-abu gray_image menggunakan fungsi np.min() Nilai minimum akan disimpan dalam variabel min_val
max_val = np.max(gray_image) # Baris ini menghitung nilai maksimum dari citra skala abu-abu gray_image menggunakan fungsi np.max() Nilai maksimum akan disimpan dalamn variabel max_val

image_cs = cv2.normalize(gray_image, image_cs, 0, 255, cv2.NORM_MINMAX) # Baris ini melakukan normalisasi pada citra skala abu-abu gray_image menggunakan fungsi cv2.normalize() dari modul cv2 (OpenCV). Citra akan dinormalisasi ke rentang nilai 0-255 dan hasilnya akan disimpan dalam variabel image_cs.

copyCamera = gray_image.copy().astype(float) # Baris ini membuat salinan dari citra skala abu-abu gray_image dengan menggunakan metode copy() dan mengubah tipe datanya menjadi float menggunakan metode astype(float). Salinan tersebut akan disimpan dalam variabel copyCamera.
output1 = copyCamera * 1.9 # baris ini akan membuat citra tersebut dikalikan dengan 1.9 dan hasilnya akan disimpan dalam variabel output1.

fig, axes = plt.subplots(5, 2, figsize=(20, 20)) # Baris ini membuat subplot grid dengan ukuran 5x2 (5 baris dan 2 kolom) dengan ukuran keseluruhan gambar 20x20 inci. Variabel fig menyimpan objek gambar (Figure), dan variabel axes menyimpan array objek sumbu (AxesSubplot) untuk setiap subplot. 
ax = axes.ravel() # ax dihasilkan dari raveling axes untuk mempermudah akses ke setiap sumbu subplot.

ax[0].imshow(image, cmap=plt.cm.gray) # Baris ini menampilkan citra input pada sumbu ax[0] dengan menggunakan imshow().
ax[0].set_title("Citra Input") #  judul "Citra Input" ditambahkan pada sumbu tersebut menggunakan set_title().
ax[1].hist(image.ravel(), bins=256) # Pada sumbu ax[1], histogram citra input ditampilkan dengan menggunakan hist() dan diatur jumlah bin sebanyak 256 menggunakan bins=256.
ax[1].set_title('Histogram Input') # Judul "Histogram Input" ditambahkan pada sumbu tersebut.

ax[2].imshow(gray_image_equalized, cmap=plt.cm.gray) # Baris ini menampilkan citra output setelah histogram equalization (HE) pada sumbu ax[2] dengan menggunakan imshow(). 
ax[2].set_title("Citra Output HE") # Judul "Citra Output HE" ditambahkan pada sumbu tersebut.
ax[3].hist(gray_image_equalized.ravel(), bins=256) # Pada sumbu ax[3], histogram citra output HE ditampilkan dengan menggunakan hist() dan diatur jumlah bin sebanyak 256 menggunakan bins=256.
ax[3].set_title('Histogram Output HE Method') # Judul "Histogram Output HE Method" ditambahkan pada sumbu tersebut.

ax[4].imshow(image_cs, cmap=plt.cm.gray) # Baris ini menampilkan citra output setelah normalisasi kontras (CS) pada sumbu ax[4] dengan menggunakan imshow(). 
ax[4].set_title("Citra Output CS") # Judul "Citra Output CS" ditambahkan pada sumbu tersebut. 
ax[5].hist(image_cs.ravel(), bins=256) # Pada sumbu ax[5], histogram citra output CS ditampilkan dengan menggunakan hist() dan diatur jumlah bin sebanyak 256 menggunakan bins=256.
ax[5].set_title('Histogram Output CS Method') # Judul "Histogram Output CS Method" ditambahkan pada sumbu tersebut.

ax[6].imshow(clahe_image, cmap=plt.cm.gray) # Baris ini menampilkan citra output setelah kontras terbatas adaptif histogram equalization (CLAHE) pada sumbu ax[6] dengan menggunakan imshow(). 
ax[6].set_title("Citra Grayscale CLAHE") # Judul "Citra Grayscale CLAHE" ditambahkan pada sumbu tersebut.
ax[7].hist(clahe_image.ravel(), bins=256) # Pada sumbu ax[7], histogram citra output CLAHE ditampilkan dengan menggunakan hist()dan diatur jumlah bin sebanyak 256 menggunakan bins=256.
ax[7].set_title('Histogram Output CLAHE Method') # Judul "Histogram Output CLAHE Method" ditambahkan pada sumbu tersebut.

ax[8].imshow(output1, cmap=plt.cm.gray) # Baris ini menampilkan citra output setelah melakukan perkalian konstanta pada sumbu ax[8] dengan menggunakan imshow().
ax[8].set_title("Citra Grayscale Perkalian Konstanta") # Judul "Citra Grayscale Perkalian Konstanta" ditambahkan pada sumbu tersebut. 
ax[9].hist(output1.ravel(), bins=256) # Pada sumbu ax[9], histogram citra output perkalian konstanta ditampilkan dengan menggunakan hist() dan diatur jumlah bin sebanyak 256 menggunakan bins=256.
ax[9].set_title('Histogram Output Perkalian Konstanta Method') # Judul "Histogram Output Perkalian Konstanta Method" ditambahkan pada sumbu tersebut.

fig.tight_layout() # Baris ini mengatur tata letak subplot agar lebih rapi menggunakan tight_layout(). 
plt.show() # gambar dan histogram ditampilkan menggunakan plt.show().
```
