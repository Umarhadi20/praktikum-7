```python
import matplotlib.pyplot as plt # Baris ini mengimpor modul matplotlib.pyplot yang digunakan untuk membuat visualisasi grafik dan plot.
%matplotlib inline

from skimage import data # Baris ini mengimpor fungsi-fungsi yang diperlukan dari modul skimage (scikit-image). data adalah submodul yang menyediakan berbagai contoh citra yang dapat digunakan untuk pengujian dan eksperimen.
from skimage.io import imread # imread adalah fungsi untuk membaca citra dari file. 
from skimage.color import rgb2gray # rgb2gray adalah fungsi untuk mengubah citra dari ruang warna RGB menjadi citra skala abu-abu (grayscale).
import numpy as np # Baris ini mengimpor modul numpy yang digunakan untuk melakukan operasi dan manipulasi data numerik, terutama dalam bentuk array.

citra1 = imread(fname="building.jpg") # Baris ini menggunakan fungsi imread dari modul skimage.io untuk membaca citra dari file. fname adalah argumen yang digunakan untuk menentukan nama file citra yang ingin dibaca. Citra pertama dibaca dari file "building.jpg"
citra2 = imread(fname="13074004.jpg") # Baris ini menggunakan fungsi imread dari modul skimage.io untuk membaca citra dari file. fname adalah argumen yang digunakan untuk menentukan nama file citra yang ingin dibaca. citra kedua dibaca dari file "13074004.jpg".

print('Shape citra 1 : ', citra1.shape) # Baris ini mencetak dimensi (shape) dari citra pertama. shape adalah atribut dari objek citra yang memberikan informasi tentang dimensi citra, yaitu jumlah baris, jumlah kolom, dan jumlah saluran warna (untuk citra berwarna).
print('Shape citra 2 : ', citra2.shape) # Baris ini mencetak dimensi (shape) dari citra kedua. shape adalah atribut dari objek citra yang memberikan informasi tentang dimensi citra, yaitu jumlah baris, jumlah kolom, dan jumlah saluran warna (untuk citra berwarna).

fig, axes = plt.subplots(1, 2, figsize=(10, 10)) # Baris ini membuat suatu figure dan satu baris dengan dua kolom subplot menggunakan subplots dari modul matplotlib.pyplot. Argumen (1, 2) menentukan jumlah baris dan kolom subplot yang ingin ditampilkan. figsize=(10, 10) mengatur ukuran gambar yang akan ditampilkan. Hasil dari plt.subplots adalah objek figure dan array axes yang berisi objek subplot.
ax = axes.ravel() # Baris ini mengubah objek axes menjadi array 1 dimensi menggunakan metode ravel(). Dalam hal ini, array axes berisi dua objek subplot yang akan digunakan.

ax[0].imshow(citra1, cmap = 'gray') # Baris ini menampilkan citra pertama dalam subplot pertama (ax[0]) menggunakan imshow. citra1 adalah citra yang ingin ditampilkan. cmap='gray' digunakan untuk menampilkan citra dalam skala abu-abu. 
ax[0].set_title("Citra 1") # set_title digunakan untuk memberikan judul "Citra 1" pada subplot pertama.
ax[1].imshow(citra2, cmap = 'gray') # Baris ini menampilkan citra kedua dalam subplot kedua (ax[1]) menggunakan imshow. citra2 adalah citra yang ingin ditampilkan. cmap='gray' digunakan untuk menampilkan citra dalam skala abu-abu. 
ax[1].set_title("Citra 2") # set_title digunakan untuk memberikan judul "Citra 2" pada subplot kedua.
```

```python
copyCitra1 = citra1.copy().astype(float) # Baris ini membuat salinan (copy) dari citra pertama (citra1) menggunakan metode copy() dari objek citra. stype(float) digunakan untuk mengkonversi tipe data citra menjadi tipe data float. Salinan citra dengan tipe data float ini akan digunakan untuk operasi pemrosesan lebih lanjut.
copyCitra2 = citra2.copy().astype(float) # # Baris ini membuat salinan (copy) dari citra kedua (citra2) menggunakan metode copy() dari objek citra. stype(float) digunakan untuk mengkonversi tipe data citra menjadi tipe data float. Salinan citra dengan tipe data float ini akan digunakan untuk operasi pemrosesan lebih lanjut.

m1, n1, _ = copyCitra1.shape # Baris ini mengambil dimensi (ukuran) citra pertama (copyCitra1) menggunakan atribut .shape. Variabel m1 dan n1 akan menyimpan ukuran baris dan kolom citra pertama
output1 = np.empty([m1, n1]) # np.empty([m1, n1])  digunakan untuk membuat array kosong dengan ukuran yang sesuai untuk menyimpan output hasil pemrosesan citra pertama (output1) Array ini akan diisi dengan nilai hasil pemrosesan pada tahap selanjutnya. Dalam hal ini, array kosong akan diisi dengan nilai float.

m2, n2, _ = copyCitra2.shape  # Baris ini mengambil dimensi (ukuran) citra citra kedua (copyCitra2) menggunakan atribut .shape. edangkan m2 dan n2 akan menyimpan ukuran baris dan kolom citra kedua.
output2 = np.empty([m2, n2]) # np.empty([m2, n2]) digunakan untuk membuat array kosong dengan ukuran yang sesuai untuk menyimpan output hasil pemrosesan citra kedua (output2). Array ini akan diisi dengan nilai hasil pemrosesan pada tahap selanjutnya. Dalam hal ini, array kosong akan diisi dengan nilai float.

print('Shape copy citra 1 : ', copyCitra1.shape) # Baris ini digunakan untuk mencetak bentuk (shape) dari salinan citra pertama (copyCitra1) Fungsi .shape digunakan untuk mengambil informasi tentang dimensi (ukuran) array atau matriks. Dalam hal ini, mencetak bentuk citra dan output akan memberikan informasi tentang jumlah baris dan kolom dari citra serta output.
print('Shape output citra 1 : ', output1.shape) # Baris ini digunakan untuk mencetak bentuk (shape) dari output citra pertama (output1). Fungsi .shape digunakan untuk mengambil informasi tentang dimensi (ukuran) array atau matriks. Dalam hal ini, mencetak bentuk citra dan output akan memberikan informasi tentang jumlah baris dan kolom dari citra serta output.

print('m1 : ',m1) # Baris ini digunakan untuk mencetak nilai dari variabel m1  yang menyimpan ukuran baris citra pertama
print('n1 : ',n1) # Baris ini digunakan untuk mencetak nilai dari variabel n1, yang menyimpan ukuran kolom citra pertama
print() # baris digunakan untuk mencetak

print('Shape copy citra 2 : ', copyCitra2.shape) # Baris ini digunakan untuk mencetak bentuk (shape) dari salinan citra kedua (copyCitra2) Fungsi .shape digunakan untuk mengambil informasi tentang dimensi (ukuran) array atau matriks. Dalam hal ini, mencetak bentuk citra dan output akan memberikan informasi tentang jumlah baris dan kolom dari citra serta output.
print('Shape output citra 3 : ', output2.shape) # # Baris ini digunakan untuk mencetak bentuk (shape) dari output citra kedua (output2). Fungsi .shape digunakan untuk mengambil informasi tentang dimensi (ukuran) array atau matriks. Dalam hal ini, mencetak bentuk citra dan output akan memberikan informasi tentang jumlah baris dan kolom dari citra serta output.
print('m2 : ',m2) # Baris ini digunakan untuk mencetak nilai dari variabel m2  yang menyimpan ukuran baris citra kedua
print('n2 : ',n2) # # Baris ini digunakan untuk mencetak nilai dari variabel n2, yang menyimpan ukuran kolom citra kedua
print() # baris ini untuk mencetak
```

```python
copyCitra1 = citra1.copy().astype(float) # Baris ini digunakan untuk membuat salinan citra pertama (citra1) dengan mengubah tipe datanya menjadi float. Menggunakan .copy() membuat salinan yang independen dari citra asli, dan .astype(float) digunakan untuk mengubah tipe data menjadi float. Salinan ini akan digunakan untuk proses filtering selanjutnya.
copyCitra2 = citra2.copy().astype(float) # Baris ini digunakan untuk membuat salinan citra kedua (citra2) dengan mengubah tipe datanya menjadi float. Menggunakan .copy() membuat salinan yang independen dari citra asli, dan .astype(float) digunakan untuk mengubah tipe data menjadi float. Salinan ini akan digunakan untuk proses filtering selanjutnya.

m1, n1 = copyCitra1.shape[:2] # Baris ini digunakan untuk mengambil ukuran baris dan kolom dari salinan citra pertama (copyCitra1) menggunakan .shape[:2] dan menyimpannya dalam variabel m1 dan n1.
output1 = np.empty([m1, n1], dtype=float) # membuat array kosong output1 dengan ukuran yang sama sebagai tempat untuk menyimpan hasil filtering. Tipe datanya juga diubah menjadi float dengan dtype=float.

# Baris ini melakukan proses filtering pada salinan citra pertama (copyCitra1) menggunakan averaging filter. Loop for digunakan untuk mengakses setiap piksel dalam citra, kecuali piksel di tepi yang tidak dapat diakses. Setiap piksel diproses dengan mengambil nilai dari piksel dan tetangganya, dan menghitung rata-ratanya. Hasil rata-rata tersebut kemudian disimpan dalam array output1.
for baris in range(1, m1-1):
    for kolom in range(1, n1-1):
        a1 = baris
        b1 = kolom
        jumlah = (
            copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1+1] +
            copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] +
            copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1]
        )
        output1[a1, b1] = (1/9) * jumlah[0]

m2, n2 = copyCitra2.shape[:2] # Baris ini digunakan untuk mengambil ukuran baris dan kolom dari salinan citra kedua (copyCitra2) menggunakan .shape[:2] dan menyimpannya dalam variabel m2 dan n2.
output2 = np.empty([m2, n2], dtype=float)  # membuat array kosong output2 untuk menyimpan hasil filtering.

# Baris ini melakukan proses filtering pada salinan citra kedua (copyCitra2) menggunakan averaging filter yang sama seperti sebelumnya. Hasil filtering disimpan dalam array output2.
for baris1 in range(1, m2-1):
    for kolom1 in range(1, n2-1):
        a1 = baris1
        b1 = kolom1
        jumlah = (
            copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1+1] +
            copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] +
            copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1]
        )
        output2[a1, b1] = (1/9) * jumlah[0]

fig, axes = plt.subplots(2, 2, figsize=(10, 10)) # Baris ini membuat sebuah subplot dengan ukuran 2x2, yang akan menampung 4 gambar. fig merupakan objek gambar utama, sedangkan axes adalah array yang berisi objek-objek sumbu (axes) yang merepresentasikan setiap subgambar dalam subplot.
ax = axes.ravel() # Baris ini meratakan array axes menjadi array 1 dimensi, sehingga mudah untuk mengakses objek sumbu (axes) individu menggunakan indeks.

ax[0].imshow(citra1, cmap='gray') # Baris ini menampilkan citra pertama (citra1) pada sumbu pertama
ax[0].set_title("Input Citra 1") # memberikan judul "Input Citra 1" pada sumbu tersebut.

ax[1].imshow(citra2, cmap='gray') # Baris ini menampilkan citra kedua (citra2) pada sumbu kedua
ax[1].set_title("Input Citra 2") # memberikan judul "Input Citra 2" pada sumbu tersebut.

ax[2].imshow(output1, cmap='gray') # Baris ini menampilkan hasil filtering citra pertama (output1) pada sumbu ketiga 
ax[2].set_title("Output Citra 1") # memberikan judul "Output Citra 1" pada sumbu tersebut.

ax[3].imshow(output2, cmap='gray') # Baris ini menampilkan hasil filtering citra kedua (output2) pada sumbu keempat
ax[3].set_title("Output Citra 2") # memberikan judul "Output Citra 2" pada sumbu tersebut.

plt.tight_layout() # Baris ini digunakan untuk mengatur tata letak subplot agar lebih rapi
plt.show() # menampilkan plot ke layar.
```
