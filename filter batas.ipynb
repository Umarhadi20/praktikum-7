```python
import matplotlib.pyplot as plt # memanggil library matplotlib.pyplot
%matplotlib inline

from skimage import data # memanggil library skimgae
from skimage.io import imread # memanggil library skimgae
from skimage.color import rgb2gray  # memanggil library skimgae
import numpy as np # memanggil library numpy

citra1 = imread(fname="kopii.jpg") # membaca file citra1
citra2 = imread(fname="kopiii.jpg") # membaca file citra2

print('Shape citra 1 : ', citra1.shape) # mencetak bentuk atau dimensi dari citra1
print('Shape citra 1 : ', citra2.shape) # mencetak bentuk dari dimensi dari citra2

fig, axes = plt.subplots(1, 2, figsize=(10, 10)) # menampilkan 1 baris subplot dan akan ada 2 subplot dalam satu baris, mengatur ukuran menjadi 10x10
ax = axes.ravel() # mengubah array dua dimensi menjadi array satu dimensi unuk mempermudah pengideksan saat mwnampilkan gambar

ax[0].imshow(citra1, cmap = 'gray') # menampilkan citra 1
ax[0].set_title("Citra 1") # judul untuk citra 1
ax[1].imshow(citra2, cmap = 'gray') # menampilkan citra 2
ax[1].set_title("Citra 2") # judul untuk citra 2
```

```python
copyCitra1 = citra1.copy() # membuat salinan citra1
copyCitra2 = citra2.copy() # membuat salinan citra2

m1, n1 = copyCitra1.shape[:2] # untuk mengambl nilai panjang dan lebar pada citra1
output1 = np.empty((m1, n1)) # tempat penyimpanan hasil operasi pada copycitra1

m2, n2 = copyCitra2.shape[:2] # untuk mengambil nilai panjang dan lebar pada citra2
output2 = np.empty((m2, n2)) # tempat penyimpanan hasil operasi pada copycitra2

print('Shape copy citra 1:', copyCitra1.shape) # mencetak bentuk atau dimensi citra1
print('Shape output citra 1:', output1.shape) # mencetak bentuk atau dimensi citra1 yang dibuat sebelumnya
print('m1:', m1) # mencetak nilai panjang dari copyCitra1
print('n1:', n1) # mencetak nilai lebar dari copyCitra1
print() # mencetak baris kosong sebagai pemisah antara citr pertama dan kedua

print('Shape copy citra 2:', copyCitra2.shape) # mencetak bentuk atau dimensi pada citra2
print('Shape output citra 2:', output2.shape) # mencetak bentuk atau dimensi dari array sebelumnya
print('m2:', m2) # mencetak nilai panjang dari copyCitra2
print('n2:', n2) # mencetak nilai lebar dari copyCitra2
print() # mencetak baris kosong
```

```python
for baris in range(0, m1-1):
    for kolom in range(0, n1-1):
        
        a1 = baris
        b1 = kolom
        
        arr = np.array([copyCitra1[a1-1, b1-1], copyCitra1[a1-1, b1], copyCitra1[a1, b1+1],
                        copyCitra1[a1, b1-1], copyCitra1[a1, b1+1], copyCitra1[a1+1, b1-1],
                        copyCitra1[a1+1, b1], copyCitra1[a1+1, b1+1]])
        
        minPiksel = np.amin(arr)
        maksPiksel = np.amax(arr)
            
        if np.any(copyCitra1[baris, kolom] < minPiksel):
            output1[baris, kolom] = minPiksel
        else:
            if np.any(copyCitra1[baris, kolom] > maksPiksel):
                output1[baris, kolom] = maksPiksel
            else:
                output1[baris, kolom] = copyCitra1[baris, kolom]


for baris1 in range(0, m2-1):
    for kolom1 in range(0, n2-1):
        
        a1 = baris1
        b1 = kolom1
        
        arr = np.array([copyCitra2[a1-1, b1-1], copyCitra2[a1-1, b1], copyCitra2[a1, b1+1],
                        copyCitra2[a1, b1-1], copyCitra2[a1, b1+1], copyCitra2[a1+1, b1-1],
                        copyCitra2[a1+1, b1], copyCitra2[a1+1, b1+1]])
        
        minPiksel = np.amin(arr)
        maksPiksel = np.amax(arr)
            
        if np.any(copyCitra2[baris1, kolom1] < minPiksel):
            output2[baris1, kolom1] = minPiksel
        else:
            if np.any(copyCitra2[baris1, kolom1] > maksPiksel):
                output2[baris1, kolom1] = maksPiksel
            else:
                output2[baris1, kolom1] = copyCitra2[baris1, kolom1]


fig, axes = plt.subplots(2, 2, figsize=(10, 10)) # membuat dua baris dan dua subplot dengan ukuran 10x10
ax = axes.ravel() # mengubah array dua dimensi menjadi array satu dimensi unuk mempermudah pengideksan saat mwnampilkan gambar

ax[0].imshow(citra1, cmap='gray') # menampilkan citra 1
ax[0].set_title("Input Citra 1") # judul input citra 1

ax[1].imshow(citra2, cmap='gray') # menampilkan citra 2
ax[1].set_title("Input Citra 2") # judul input citra 2

ax[2].imshow(output1, cmap='gray') # menampilkan output citra 1
ax[2].set_title("Output Citra 1") # judul output citra 1

ax[3].imshow(output2, cmap='gray') # menampilkan output citra 2
ax[3].set_title("Output Citra 2") # judul output citra 2

plt.show() # menampilkan gambar
```
